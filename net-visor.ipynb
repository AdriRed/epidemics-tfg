```python
import networkx as nx
g = nx.read_weighted_edgelist("./nets/out.moreno_beach_beach", delimiter=" ",comments='%',nodetype=int)
```

```python
import matplotlib.pyplot as plt

degrees = dict(g.degree())
weights = {}

for a, b, data in g.edges(data=True):
    weight = data['weight']
    if (weight not in weights):
        weights[weight] = 0
    weights[weight]+=1

plt.scatter(degrees.keys(), degrees.values(), label='degrees')
plt.scatter(weights.keys(), weights.values(), label='weight')
plt.legend()
plt.show()
```
